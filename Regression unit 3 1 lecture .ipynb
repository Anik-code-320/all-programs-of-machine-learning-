```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
```

```python
import pandas as pd 
dataset=pd.read_csv(url,na_values='?',skipinitialspace=True,header=None,sep='\s+')
```

```python
dataset.head()
```

```python
columns=['mpg', 'cylinders','dosplacement','horsepower','weight','acceleration','year','origin' ,'car_name']
```

```python
dataset.columns=columns
```

```python
dataset.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
sns.pairplot(dataset,height=1)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['car_name']=le.fit_transform(dataset['car_name'])
```

```python
import numpy as np
```

```python
cm=np.corrcoef(dataset[columns].values.T)
```

```python
hm=sns.heatmap(cm,annot=True)
```

```python
y=dataset[[0]].values
x=dataset.drop(columns=[0, 8])
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3)
```

```python
lr.fit(x_train, y_train)
```

```python
from sklearn.impute import SimpleImputer
im=SimpleImputer(strategy='most_frequent')
dataset=im.fit_transform(dataset)
```

```python
pd.DataFrame(dataset).isnull().sum()
```

```python
dataset.head()
```

```python
train_predict=lr.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print(mean_squared_error(train_predict, y_train))
print(r2_score(train_predict, y_train))
```

```python
test_predict=lr.predict(x_test)
```

```python
print(mean_squared_error(test_predict, y_test))
print(r2_score(test_predict, y_test))
```
