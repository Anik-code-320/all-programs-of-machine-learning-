```python
#perceptron prediction

import numpy as np
weight=np.array([0,1,0])
bias=0
input1=np.array([1,-1,1])
def predict(input, weight ,bias):
    y_net=np.sum(input*weight)+bias
    if y_net>0:
        print(" 1, it is an apple")
        
    else:
        print(" 1, it is an orange")
        
predict(input1,weight,bias)
        
```
